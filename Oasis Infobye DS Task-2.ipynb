```python
# Plotting the distribution of Estimated Unemployment Rate (%) for both datasets
plt.figure(figsize=(10, 6))
sns.histplot(data=df1, x=' Estimated Unemployment Rate (%)', bins=20, kde=True, color='blue', label='Dataset 1')
sns.histplot(data=df2, x=' Estimated Unemployment Rate (%)', bins=20, kde=True, color='red', label='Dataset 2')
plt.title('Distribution of Estimated Unemployment Rate (%)')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Plotting the distribution of Estimated Unemployment Rate (%) for both datasets
plt.figure(figsize=(10, 6))
sns.lineplot(data=df1, x=' Date', y=' Estimated Labour Participation Rate (%)', color='blue', label='Dataset 1')
sns.lineplot(data=df2, x=' Date', y=' Estimated Labour Participation Rate (%)', color='red', label='Dataset 2')
plt.title('Estimated Labour Participation Rate (%) Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.legend()
plt.show()
```
